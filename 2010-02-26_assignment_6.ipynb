```python
pd.set_option('display.max_columns', None)
import pandas as pd
df1=pd.read_csv('C:\\Users\\prash\\data\\data\\Medicare_Hospital_Spending_by_Claim.csv')
```

```python
df1.drop(columns=['Measure Start Date','Measure End Date'], inplace=True)
```

```python
df1.columns = ['Hospital', 'Provider ID', 'State', 'Period', 'Claim Type', 'Avg Spending Hospital', 
               'Avg Spending State', 'Avg Spending Nation', 
               'Percent Spending Hospital', 'Percent Spending State', 'Percent Spending Nation']
```

```python
import warnings
warnings.filterwarnings("ignore")

df2=df1.loc[df1['State'] == 'TX',]
repl={"$":"", "\%":""}
col_nm=['Avg Spending Hospital', 'Avg Spending State', 'Avg Spending Nation']
for col in col_nm:
    df2[col] = df2[col].str.replace('$','').astype(float)

col_nm=['Percent Spending Hospital', 'Percent Spending State','Percent Spending Nation']
for col in col_nm:
    df2[col] = df2[col].str.replace('\%','').astype(float)
```

```python
df2[['Hospital','Claim Type','Avg Spending Hospital','Avg Spending State','Avg Spending Nation']]. \
groupby(['Hospital','Claim Type']).aggregate('mean')
```

```python
df2[['Hospital','Claim Type','Percent Spending Hospital', 'Percent Spending State','Percent Spending Nation']]. \
groupby(['Hospital','Claim Type']).aggregate(['mean','max'])
```
